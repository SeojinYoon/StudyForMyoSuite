# Common Libraries

```python
import os
os.environ["MUJOCO_GL"] = "egl"

import sys
import skvideo.io
import numpy as np
from myosuite.utils import gym
```

# Custom Libraries

```python
import sys
sys.path.append("/home/seojin/Seojin_commonTool/Module")

from video_tool import save_video
from jupyter_util import show_video
```

# Environment

```python
env = gym.make('myoHandPoseRandom-v0')
```

# Simulation

```python
env.reset()
frames = []
for _ in range(100):
    frame = env.sim.renderer.render_offscreen(width=400,
                                              height=400,
                                              camera_id=0)
    frames.append(frame)
    env.step(env.action_space.sample())
env.close()
```

# Save video

```python
video_dir_path = os.path.join("/mnt/ext1/seojin/temp")
video_path = os.path.join(video_dir_path, "temp.mp4")
skvideo.io.vwrite(video_path, np.asarray(frames),outputdict={"-pix_fmt": "yuv420p"})
print(f"save: {video_path}")
```

# Show video

```python
show_video(video_path)
```
