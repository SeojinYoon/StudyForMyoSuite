```python
import os, sys, shutil, mediapy
import numpy as np

if shutil.which('nvidia-smi') is not None:
    os.environ['MUJOCO_GL'] = 'egl'
import mujoco
```

```python
model_path = "/Users/seojin/O2MConverter/models/converted/MoBL_ARMS_model_for_mujoco_converted/MoBL_ARMS_model_for_mujoco_converted.xml"
geometry_path = "/Users/seojin/O2MConverter/models/converted/MoBL_ARMS_model_for_mujoco_converted/Geometry"
```

```python
model = mujoco.MjModel.from_xml_path(model_path)
data = mujoco.MjData(model)

cam = mujoco.MjvCamera()
with mujoco.Renderer(model) as renderer:
    mujoco.mj_forward(model, data)

    cam.azimuth = 45
    cam.elevation = -20
    cam.distance = 1.7
    cam.lookat = np.array([0, 0, 0.5])
    renderer.update_scene(data, camera=cam)

    mediapy.show_image(renderer.render())
```
