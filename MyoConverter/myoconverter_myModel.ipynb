# Common Libraries

```python
import os, sys

# myoconverter
if 'CONDA_PREFIX' not in os.environ:
    os.environ['CONDA_PREFIX'] = sys.prefix
myoconverter_path = "/home/seojin/myoconverter"
sys.path.append(myoconverter_path)
from myoconverter.O2MPipeline import O2MPipeline
import xml.etree.ElementTree as ET
from pathlib import Path
```

# Custom Libraries

```python
def extract_muscle_names(osim_file_path, tag = "Millard2012EquilibriumMuscle"):
    tree = ET.parse(osim_file_path)
    root = tree.getroot()

    muscle_names = []
    for muscle in root.iter(tag):
        name = muscle.get('name')
        if name:
            muscle_names.append(name)

    return muscle_names
```

# Paths

```python
osim_dir_path = "/mnt/sdb2/DeepProprioception/Projects/DP01_mri/Opensim/Model"
osim_file_path = os.path.join(osim_dir_path, "scaled_model.osim")
geometry_dir_path = os.path.join(osim_dir_path, "Geometry")
model_name = Path(osim_file_path).stem
output_dir_path = f"/mnt/ext1/seojin/temp/{model_name}"
print(output_dir_path)
os.makedirs(output_dir_path, exist_ok = True)
os.system(f"rm -r {output_dir_path}/*")
```

```python
osim_model_path = Path(osim_file_path)
parser = ET.XMLParser(target=ET.TreeBuilder(insert_comments=True))
tree = ET.parse(osim_model_path, parser)
root = tree.getroot()
coords = root.findall(".//Coordinate")

unlock_coords = ["elv_angle", "shoulder_elv", "shoulder_rot", "elbow_flexion", "pro_sup", "deviation", "flexion"]
for i in range(len(coords)):
    name = coords[i].get("name")
    if name in unlock_coords:
        lock_element = coords[i].find("locked")
        lock_element.text = "false"

lock_coords = ["r_x", "r_y", "r_z", "t_x", "t_y", "t_z"]
for i in range(len(coords)):
    name = coords[i].get("name")
    if name in lock_coords:
        lock_element = coords[i].find("locked")
        lock_element.text = "true"

tree.write(osim_model_path, xml_declaration=True, encoding='UTF-8')
```

# Params

```python
# define pipeline configurations
kwargs = {}  # define kwargs inputs
kwargs['convert_steps'] = [1,2,3]    # All three steps selected
muscle_list = extract_muscle_names(osim_file_path)
kwargs['muscle_list'] = muscle_list           # No specific muscle selected, optimize all of them

kwargs['osim_data_overwrite'] = True   # Overwrite the Osim model state files
kwargs['conversion'] = True            # Yes, perform 'Cvt#' process
kwargs['validation'] = False            # Yes, perform 'Vlt#' process
kwargs['speedy'] = False               # Do not reduce the checking notes to increase speed
kwargs['generate_pdf'] = True          # Do not generate validation pdf report
kwargs['add_ground_geom'] = False       # Add ground to the model
kwargs['treat_as_normal_path_point'] = False    # Using constraints to represent moving and conditional path points
```

# Execution

```python
%%capture pipeline_output

O2MPipeline(osim_file_path, geometry_dir_path, output_dir_path, **kwargs)
```
