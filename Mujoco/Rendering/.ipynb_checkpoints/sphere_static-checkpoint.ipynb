# Common Libraries

```python
import os
import shutil
if shutil.which('nvidia-smi') is not None:
    os.environ['MUJOCO_GL'] = 'egl'

import mujoco
import mediapy as media
```

# Model specification

```python
xml = """
<mujoco>
  <worldbody>
    <geom name="red_box" type="box" size=".2 .2 .2" rgba="1 0 0 1"/>
    <geom name="green_sphere" pos=".2 .2 .2" size=".1" rgba="0 1 0 1"/>
  </worldbody>
</mujoco>
"""
```

# Initialize model & data

```python
model = mujoco.MjModel.from_xml_string(xml) # model of the system, such as the world, bodies, joints, actuators, etc.
data = mujoco.MjData(model) # data related to the model, such as positions, velocities, forces, etc.
```

# Initial rendering

```python
# Make renderer, render and show the pixels
with mujoco.Renderer(model) as renderer:
  mujoco.mj_forward(model, data)
  renderer.update_scene(data)

  media.show_image(renderer.render())
```

# Rendering sphere with light

```python
xml = """
<mujoco>
  <worldbody>
    <light name="top" pos="0 0 2"/>
    <geom name="red_box" type="box" size=".2 .2 .2" rgba="1 0 0 1"/>
    <geom name="green_sphere" pos=".2 .2 .2" size=".1" rgba="0 1 0 1"/>
  </worldbody>
</mujoco>
"""
model = mujoco.MjModel.from_xml_string(xml)
data = mujoco.MjData(model)

with mujoco.Renderer(model) as renderer:
  mujoco.mj_forward(model, data)
  renderer.update_scene(data)

  media.show_image(renderer.render())
```
